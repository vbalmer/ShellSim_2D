```python
import numpy as np

Dmh = np.random.randn(3,3)
Dmbh = np.random.randn(3,3)
Dsh = np.random.randn(2,2)
Dbh = np.random.randn(3,3)

print(Dmh)
print(Dmbh)
print(Dbh)
print(Dsh)
```

```python
De_1 = np.hstack([Dmh,Dmbh,np.zeros((3, 2))])
De_2 = np.hstack([Dmbh,Dbh,np.zeros((3,2))])
De_3 = np.hstack([np.zeros((2,3)), np.zeros((2,3)), Dsh])
De = np.vstack([De_1, De_2, De_3])

print(De_1)
print(De_2)
print(De_3)
print(De)
```
